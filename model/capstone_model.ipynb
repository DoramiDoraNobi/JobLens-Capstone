```python
collumn_drop = ['job_link', 'last_processed_time', 'got_summary', 'got_ner', 'is_being_worked', 'company', 'job_location', 'first_seen',
                'search_city','search_country', 'search_position', 'job_level', 'job_type']
df = merge_clean_df.drop(collumn_drop, axis = 1)
df.head()
```

```python
df_sam = df.sample(n= 10000, random_state= 42)
df_sam
```

text vectorization

```python
min_df = 1

vectorizer = CountVectorizer(min_df = min_df)
vectorizer_data = vectorizer.fit_transform(df_sam['job_skills'])
vectorizer_data.toarray()
```

```python
vectorizer_data = pd.DataFrame(vectorizer_data.toarray(), columns = vectorizer.get_feature_names_out())
vectorizer_data
```

```python
tfid = TfidfVectorizer(min_df= min_df)
tfid_data = tfid.fit_transform(df_sam['job_skills'])
tfid_data.toarray()
```

```python
tfid_data = pd.DataFrame(tfid_data.toarray(), columns = tfid.get_feature_names_out())
tfid_data
```
